# PLOTTING HISTOGRAM 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style('whitegrid')

# Plotting the distribution of ages
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Gender distribution
print('Gender distribution:\
', df['ptgender'].value_counts())

# Diagnosis distribution
print('Diagnosis distribution:\
', df['dx'].value_counts())
```

The age distribution in the dataset shows a wide range, with a slight skew towards older ages, indicating a majority of participants are in the older age brackets. The gender distribution is relatively balanced with 461 males and 375 females.

The diagnosis distribution is as follows:

AD (Alzheimer's Disease): 361 cases
MCI (Mild Cognitive Impairment): 244 cases
NL (No Listed impairment): 231 cases

This indicates a significant portion of the dataset consists of individuals diagnosed with Alzheimer's Disease, followed by those with Mild Cognitive Impairment and a healthy control group.

```python


# Plotting the distribution of initial diagnoses (dxbl)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='dxbl', order = df['dxbl'].value_counts().index)
plt.title('Distribution of Initial Diagnoses (dxbl)')
plt.xlabel('Diagnosis')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting the distribution of follow-up diagnoses (dx)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='dx', order = df['dx'].value_counts().index)
plt.title('Distribution of Follow-up Diagnoses (dx)')
plt.xlabel('Diagnosis')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

These plots visualize the frequency of each diagnosis at the baseline (dxbl) and at follow-up (dx), providing insights into how the distribution of diagnoses changes over time within the dataset.

```python
# Grouping the data by diagnosis and month, then calculating the mean of cognitive test scores
grouped_data = df.groupby(['dx', 'month'])[['cdrsb.bl', 'adas11.bl', 'adas13.bl', 'mmse.bl']].mean().reset_index()

# Plotting the trends of cognitive test scores over time for different diagnoses
fig, axs = plt.subplots(2, 2, figsize=(20, 12))

# Setting the titles for each subplot
axs[0, 0].set_title('CDRSB Score Over Time')
axs[0, 1].set_title('ADAS11 Score Over Time')
axs[1, 0].set_title('ADAS13 Score Over Time')
axs[1, 1].set_title('MMSE Score Over Time')

# Plotting each cognitive test score
sns.lineplot(data=grouped_data, x='month', y='cdrsb.bl', hue='dx', ax=axs[0, 0])
sns.lineplot(data=grouped_data, x='month', y='adas11.bl', hue='dx', ax=axs[0, 1])
sns.lineplot(data=grouped_data, x='month', y='adas13.bl', hue='dx', ax=axs[1, 0])
sns.lineplot(data=grouped_data, x='month', y='mmse.bl', hue='dx', ax=axs[1, 1])

# Adjusting the layout
plt.tight_layout()
plt.show()
```

This plot visualizes the trends of cognitive test scores (cdrsb.bl, adas11.bl, adas13.bl, mmse.bl) over time for different diagnoses. It provides insights into how cognitive functions, as measured by these scores, change across different stages of Alzheimer's Disease and related conditions over the course of the study.

# ANALYSING THE CORRELATION FOR NUMERICAL COLUMNS

```python
# Drawing pairplots for a subset of attributes to visualize pairwise relationships
# Selecting a subset of numerical columns for clarity in visualization
subset_cols = ['age', 'pteducat', 'mmse.bl', 'cdrsb.bl', 'adas11.bl']
sns.pairplot(df[subset_cols])
plt.suptitle('Pairplot of Selected Attributes', y=1.02)
plt.show()
```

The pairplot offered a glimpse into the pairwise relationships and distributions, facilitating the identification of potential correlations or patterns among the selected variables.

```python
selected_cols = df[['cdrsb.bl', 'adas11.bl']]
for column in selected_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df[column])
    plt.xticks(rotation=90)
    plt.title('Boxplot of Numerical Attributes')
    plt.show()
```

```python
import numpy as np

# Selecting numerical columns for correlation analysis
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Calculating correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Plotting the heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

The color scale, ranging from blue (negative correlation) to red (positive correlation), helps in identifying the strength and direction of the relationships between variables. Values closer to 1 or -1 indicate a strong positive or negative correlation, respectively, while values around 0 suggest little to no linear correlation.

This visualization helps to uncover potential relationships between variables that could be significant for further analysis or predictive modeling.

```python
 #Calculating Pearson correlation for the subset of attributes
pearson_corr = df[subset_cols].corr()
print(pearson_corr)
```

```python
# calculate Pearson's correlation
from scipy.stats import pearsonr
pearson_correlations_subset = {}

# Loop through all possible pairs of the subset columns
for i in range(len(subset_cols)):
    for j in range(i+1, len(subset_cols)):
        col1 = subset_cols[i]
        col2 = subset_cols[j]
        # Calculate Pearson's correlation
        corr, _ = pearsonr(df[col1], df[col2])
        # Store the correlation in the dictionary
        pearson_correlations_subset[col1 + ' & ' + col2] = corr

# Convert the dictionary to a DataFrame for better visualization
pearson_df_subset = pd.DataFrame(list(pearson_correlations_subset.items()), columns=['Column Pair', 'Pearson Correlation'])
print(pearson_df_subset)
```

CDRSB and ADAS11 scores have a strong positive correlation (), meaning that as the severity of dementia symptoms increases, the cognitive deficits measured by ADAS11 also increase.
